```python
import math
from scipy.stats import entropy
def get_klpq_div(p_probs, q_probs):
    kl_div = 0.0
    
    for pi, qi in zip(p_probs, q_probs):
        kl_div += pi*np.log(pi/qi)
    
    return kl_div

discrete_scene1_explicit_array_mean  = np.zeros((1,256))[0]
for i in scene1_explicit_array_mean:
    discrete_scene1_explicit_array_mean[math.floor(i)] +=1
prob_q = discrete_scene1_explicit_array_mean/np.sum(discrete_scene1_explicit_array_mean)

print(prob_q)

discrete_scene1_dqn_array_mean = np.zeros((1,256))[0]
for i in scene1_dqn_array_mean:
    discrete_scene1_dqn_array_mean[math.floor(i)] +=1
prob_p = discrete_scene1_dqn_array_mean/np.sum(discrete_scene1_dqn_array_mean)

print(entropy(prob_p,prob_q))

```
